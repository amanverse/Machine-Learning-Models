```python
#import modules
import pandas as pd
import os 
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import load_img   
```

```python
# creating  label data for training
y_train=[]
for class_ in os.listdir('/content/training_set'):
  for img in os.listdir('/content/training_set/'+class_):
    if class_ == 'cats':
      y_train.append('Cat')
    else :
     y_train.append("Dog") 
  
print(y_train)
```

```python
# creating label data for test
y_test=[]
for class_ in os.listdir('/content/test_set'):
  for img in os.listdir('/content/test_set/'+class_):
          if class_=='cats':
            y_test.append ('Cat') 
          else:
            y_test.append('Dog')

print(y_test)
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_gen= ImageDataGenerator(rescale=1./255, # normalisation of images
                            shear_range= 0.2,
                              zoom_range= 0.2,
                              horizontal_flip= True,
                              fill_mode = 'nearest')
test_gen= ImageDataGenerator(rescale=1./255)

x_train = train_gen.flow_from_directory('/content/training_set',
                                                target_size=(100,100),
                                                batch_size= 16,
                                                class_mode= 'categorical')

x_test= test_gen.flow_from_directory('/content/test_set',
                                      target_size =(100,100),
                                      batch_size = 16,
                                      class_mode='categorical')                                                    
                                              

```

```python
# create model
from keras import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense
model = Sequential([Conv2D (16,(3,3),activation = 'relu', input_shape=(100,100,3)),
                   MaxPool2D ((2,2)),
                   Conv2D (32,(3,3),activation='relu'),
                   MaxPool2D (2,2),
                   Conv2D (64,(3,3),activation='relu'),
                   MaxPool2D (2,2),
                   Flatten(),
                    Dense (512,activation='relu'),
                    Dense (1, activation='relu')])                    
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) 
```

```python
model.summary()
```

```python
# model fit
Cnn=model.fit(x_train, epochs=10, validation_data=x_test)
```
