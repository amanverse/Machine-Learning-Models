```python
#import modules
import pandas as pd
import os 
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import load_img   
```

```python
# creating data for training
x_train=[]
y_train=[]
for class_ in os.listdir('/content/training_set'):
  for img in os.listdir('/content/training_set/'+class_):
    if class_ == 'cats':
      y_train.append('Cat')
    else :
     y_train.append("Dog") 
    x_train.append(os.path.join(class_,img))
print(x_train[0],y_train[0])
```

```python
import random
a=list(zip(x_train,y_train))
random.shuffle(a)
x_train,y_train=zip(*a)
print(x_train)
print(y_train)
# image's name and lable names match
```

```python
# creating data for test
x_test=[]
y_test=[]
for class_ in os.listdir('/content/test_set'):
  for img in os.listdir('/content/test_set/'+class_):
          if class_=='cats':
            y_test.append ('Cat') 
          else:
            y_test.append('Dog')
          x_test.append(os.path.join(class_,img))    
```

```python
b=list(zip(x_test,y_test))
random.shuffle(b)
x_test,y_test=(zip(*b))
print(x_test)
print(y_test)
```
